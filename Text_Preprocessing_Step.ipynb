```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
def remove_stopword(text):
    new_text=[]
    for word in text.split():
        if word not in stopwords.words("english"):
            new_text.append(word)
        else:
            pass
    return " ".join(new_text)
        
```

```python
def remove_emoji(text):
    pattern=re.compile("["
                       u"\U0001F600-\U0001F64F"  # Emoticons 😀-😏
                       u"\U0001F300-\U0001F5FF"  # Symbols & pictographs 🌍-🎏
                       u"\U0001F680-\U0001F6FF"  # Transport & map symbols 🚀-🚦
                       u"\U0001F1E0-\U0001F1FF"  # Flags (iOS) 🇦-🇿
                       u"\U00002702-\U000027B0"  # Dingbats ✂-➰
                       u"\U000024C2-\U0001F251"  # Enclosed characters Ⓜ-🉑
                       "]+",flags=re.UNICODE
                      )
    return pattern.sub(r"",text)
```

```python
paragraph=remove_emoji(paragraph)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
PS=PorterStemmer()
```

```python
def word_stemmer(text):
    stem_word=[]
    for word in text:
        stem_word.append(PS.stem(word))
    return " ".join(stem_word)
```

```python
word_stemmer(token_para)
```

```python
# lemmetizer used when you want to show output to user 
# when matter is related speed go with stemming
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
WNL=WordNetLemmatizer()
```

```python
def lemmetization(text):
    lem_word=[]
    for word in text:
        lem_word.append(WNL.lemmatize(word,pos="v"))
    return " ".join(lem_word)
```

```python
lemmetization(token_para)
```
