```python
import numpy as np
import pandas as pd
```

```python
Dict={"Text":["people watch Vidyantra","Vidyantra watch Vidyantra","people write Vidyantra","Vidyantra write comments"],"output":[1,1,0,0]}
df=pd.DataFrame(Dict)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df["Text"])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
print(cv.transform(["Vidyantra watch the write comments"]).toarray())
```
